```python
# Creating Histogram and Printing Values
import numpy as np

d = np.random.randint(0,225,(128,128),dtype='uint8')
floatD = d.astype('float')
floatD = floatD / 255

def get_histogram(array):
    #bins = 100
    aMin = np.round(array.min())
    aMax = np.round(array.max())
    edge = np.linspace(aMin, aMax, 100+1)
    histogram = np.zeros((100))
    
    for i in range(100):
        histogram[i] = np.sum(np.logical_and(array >= edge[i], edge[i+1] > array))
    histogram[-1] += np.sum(array == edge[-1])
    edge = (edge[:-1] + edge[1:]) / 2
    edge = edge.astype('float')
    histogram = histogram.astype('int')
    
    return tuple((edge, histogram))

edge, histogram = get_histogram(floatD)

print('Edge: \n', edge)
print('Histogram: \n', histogram)
assert np.sum(histogram) == floatD.shape[0]*floatD.shape[1]


#References: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
#            https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
#            https://www.w3schools.com/python/python_howto_reverse_string.asp
#            https://www.codingem.com/reverse-slicing-in-python/
#            https://flexiple.com/python/python-new-line/
```

```python
#Inputting values for Histogram and Displaying it

import matplotlib.pyplot as plt

def plot_histogram(array, title = 'My Histogram', color='red', ec='orange'): 

    plt.figure(figsize=(15,10))

    edge, histogram = get_histogram(array)
    
    W = 0.5 * (edge[1] - edge[0])
    
    
    plt.title(title)
    plt.bar(edge, histogram, width = W, color = color , label = 'Array histogram')
    plt.legend()
    plt.xlabel('Edges')
    plt.ylabel('Frequency')


    
plot_histogram(d)

#References: https://www.statology.org/matplotlib-histogram-color/
#            https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm#:~:text=bar()%20function%20allows%20you,()%20plots%20the%20blue%20bars.
#            https://stackoverflow.com/questions/33203645/how-to-plot-a-histogram-using-matplotlib-in-python-with-a-list-of-data
#            https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar
#            https://datatofish.com/bar-chart-python-matplotlib/
```

```python
# New Function and Displaying new Image

colors = ['Red', 'Green', 'Blue']
def plot_RGB(array): 
    plt.figure(figsize=(15, 10))
    plt.subplot(1,2,1)
    plt.title('Colored Image')
    plt.imshow(color_IMG)
    
    plt.subplot(1,2,2)
    plt.title('RGB Histogram')

    
    for i in range(3):
        edge, histogram = get_histogram(array[:,:,i])
        plt.plot(edge, histogram, label=f'Channel {colors[i]}', color=colors[i].lower())

    plt.xlabel('Color/Light Intensity')
    plt.ylabel('Color Frequency')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    plt.close()
    
plot_RGB(color_IMG)
```

```python
#Creating Contour algorithm and displaying its effect on the image

gray_IMG = cv2.cvtColor(color_IMG, cv2.COLOR_BGR2GRAY)
threshold, bMASK = cv2.threshold(gray_IMG, 130, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(image = bMASK, mode = cv2.RETR_TREE, method = cv2.CHAIN_APPROX_NONE)
copy_IMG = color_IMG.copy()
cv2.drawContours(image=copy_IMG, contours=contours, contourIdx = -1, color = (0, 255, 0), thickness = 1, lineType = cv2.LINE_AA)
copy_RGB_IMG = cv2.cvtColor(copy_IMG, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (25, 15))
plt.imshow(copy_RGB_IMG)
#plt.axis('off')

#References: https://learnopencv.com/contour-detection-using-opencv-python-c/
```
